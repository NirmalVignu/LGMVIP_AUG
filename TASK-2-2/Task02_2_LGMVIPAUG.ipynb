```python
df1=df.drop('PetalLengthCm',axis=1)
```

```python
df1.head()
```

# Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Species']=le.fit_transform(df1['Species'])
```

```python
df1['Species'].unique()
```

# Train Test Split

```python
X=df1.drop('Species',axis=1).values
Y=df1['Species']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=2,stratify=Y)
```

# Trainig Our Machine Learning Model

```python
from sklearn.tree import DecisionTreeClassifier 
clf=DecisionTreeClassifier(random_state=100)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn import tree
tree.plot_tree(clf)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
cm=np.array(confusion_matrix(y_test,y_pred))
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
import seaborn as sns
sns.heatmap(cm/np.sum(cm), annot=True, fmt='.2%', cmap='Blues')
```

```python
print(classification_report(y_test,y_pred))
```
